```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#load the data
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

```python
#loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
#Adding the target column now
data_frame['label'] = breast_cancer_dataset.target
```

1 -> Benign (Non - Cancerous), 0 -> Malignant (Cancerous)

```python
data_frame.groupby('label').mean()
```

```python
#Seperating the features and target
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
x
```

```python
y
```

```python
#splitting the data now
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
x.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
print(x_train_std)
```

Looking into the neural network

image.png

```python
#import the libraries
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural network
#keras.Sequential is the simplest way to build a neural network in Keras.
#It lets you stack layers one after another in a straight line.

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #input layer
    keras.layers.Dense(20, activation='relu'), # Hidden layer #It outputs the input directly if positive and zero if negative,
    keras.layers.Dense(2, activation='sigmoid') # Output layer

])
```

```python
#compiling the neural network
model.compile(
    optimizer="adam", #The optimizer decides how big those adjustments should be., Adam stands for Adaptive Moment Estimation.
    loss="sparse_categorical_crossentropy", #Loss function measures how wrong the model is.
    metrics=["accuracy"] #Metrics are used to monitor model performance.
)
```

```python
#training the neural network
history = model.fit(x_train, y_train, validation_split=0.1, epochs=10)
```

```python
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

Visualizing the Accuracy and LOSS

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
```
